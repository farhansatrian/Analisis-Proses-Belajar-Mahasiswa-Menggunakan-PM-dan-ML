## Cell 11 — Gabungkan dua log mentah dengan penanda kelas

```python
log_02_raw = log_02.copy()
log_03_raw = log_03.copy()

log_02_raw["kelas"] = "IF-49-02"
log_03_raw["kelas"] = "IF-49-03"

log_raw = pd.concat([log_02_raw, log_03_raw], ignore_index=True)

print("Ukuran log gabungan:", log_raw.shape)
display(log_raw.head())
```

## Cell 18 — Ubah Time menjadi datetime untuk analisis waktu

```python
log_raw["Time_parsed"] = pd.to_datetime(
    log_raw["Time"],
    format="%d/%m/%y, %H:%M:%S",
    errors="coerce"
)

print("Tipe data Time_parsed:")
print(log_raw["Time_parsed"].dtype)

print("\nJumlah Time yang gagal dibaca:")
print(log_raw["Time_parsed"].isna().sum())

print("\nTimestamp awal:")
print(log_raw["Time_parsed"].min())

print("\nTimestamp akhir:")
print(log_raw["Time_parsed"].max())
```

## Cell 19 — Cek missing value

```python
missing_values = log_raw.isna().sum().sort_values(ascending=False)

display(missing_values)
```

## Cell 20 — Buat tabel ringkasan raw log

```python
summary_raw_log = (
    log_raw
    .groupby("kelas")
    .agg(
        jumlah_event=("Event name", "count"),
        jumlah_user_unik=("User full name", "nunique"),
        jumlah_event_name_unik=("Event name", "nunique"),
        jumlah_component_unik=("Component", "nunique"),
        timestamp_awal=("Time_parsed", "min"),
        timestamp_akhir=("Time_parsed", "max")
    )
    .reset_index()
)

summary_gabungan = pd.DataFrame({
    "kelas": ["Gabungan"],
    "jumlah_event": [len(log_raw)],
    "jumlah_user_unik": [log_raw["User full name"].nunique()],
    "jumlah_event_name_unik": [log_raw["Event name"].nunique()],
    "jumlah_component_unik": [log_raw["Component"].nunique()],
    "timestamp_awal": [log_raw["Time_parsed"].min()],
    "timestamp_akhir": [log_raw["Time_parsed"].max()]
})

summary_raw_log = pd.concat(
    [summary_raw_log, summary_gabungan],
    ignore_index=True
)

display(summary_raw_log)
```

## Cell 21 — Buat tabel top event name dengan persentase

```python
top_event_name = (
    log_raw["Event name"]
    .value_counts()
    .reset_index()
)

top_event_name.columns = ["event_name", "jumlah_event"]
top_event_name["persentase"] = (
    top_event_name["jumlah_event"] / len(log_raw) * 100
).round(2)

display(top_event_name.head(20))
```

## Cell 22 — Buat tabel top component dengan persentase

```python
top_component = (
    log_raw["Component"]
    .value_counts()
    .reset_index()
)

top_component.columns = ["component", "jumlah_event"]
top_component["persentase"] = (
    top_component["jumlah_event"] / len(log_raw) * 100
).round(2)

display(top_component.head(20))
```

## Cell 23 — Simpan tabel ke outputs/tables

```python
TABLES_DIR = PROJECT_ROOT / "outputs" / "tables"
TABLES_DIR.mkdir(parents=True, exist_ok=True)

summary_raw_log.to_csv(TABLES_DIR / "01_summary_raw_log.csv", index=False)
top_event_name.to_csv(TABLES_DIR / "01_top_event_name_raw.csv", index=False)
top_component.to_csv(TABLES_DIR / "01_top_component_raw.csv", index=False)

print("Tabel berhasil disimpan ke:", TABLES_DIR)
```

## Kesimpulan Data Understanding

Pada tahap data understanding, dua file log LMS dari kelas IF-49-02 dan IF-49-03 berhasil dibaca dan digabungkan. Total data mentah terdiri dari 165.646 event dengan 10 kolom setelah penambahan atribut kelas. Kolom utama yang tersedia meliputi Time, User full name, Event context, Component, Event name, Description, Origin, dan IP address.

Hasil pemeriksaan awal menunjukkan bahwa terdapat 96 user unik pada raw log gabungan. Jumlah ini lebih besar dibandingkan jumlah mahasiswa pada data nilai resmi, sehingga raw log masih mengandung aktivitas dari user selain mahasiswa, seperti dosen, admin, atau aktivitas sistem. Oleh karena itu, tahap preprocessing berikutnya perlu melakukan penyaringan user berdasarkan daftar mahasiswa pada data nilai.

Distribusi aktivitas menunjukkan bahwa raw log didominasi oleh aktivitas kuis. Event name dengan frekuensi tertinggi adalah Quiz attempt viewed dan Quiz attempt updated. Dari sisi Component, komponen Quiz juga menjadi komponen paling dominan. Hal ini menunjukkan bahwa interaksi mahasiswa pada LMS banyak terjadi pada aktivitas evaluasi atau pengerjaan kuis.

Kolom Time berhasil dikonversi menjadi format datetime tanpa nilai gagal parsing. Selain itu, kolom utama yang diperlukan untuk process mining, yaitu User full name, Event name, Component, dan Time, tidak memiliki missing value. Dengan demikian, data mentah memiliki struktur yang memadai untuk dilanjutkan ke tahap persiapan label dan preprocessing event log.
