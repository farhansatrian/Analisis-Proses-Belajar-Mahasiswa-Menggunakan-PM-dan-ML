# Cell 1 — Import library

```python
import pandas as pd
import numpy as np
from pathlib import Path
```

# Cell 2 — Tentukan folder project

```python
PROJECT_ROOT = Path.cwd()

if PROJECT_ROOT.name == "notebooks":
    PROJECT_ROOT = PROJECT_ROOT.parent

DATA_DIR = PROJECT_ROOT / "data"
PROCESSED_DIR = DATA_DIR / "processed"
OUTPUT_TABLES_DIR = PROJECT_ROOT / "outputs" / "tables"

PROCESSED_DIR.mkdir(parents=True, exist_ok=True)
OUTPUT_TABLES_DIR.mkdir(parents=True, exist_ok=True)

print("Project root:", PROJECT_ROOT)
print("Processed dir:", PROCESSED_DIR)
print("Output tables dir:", OUTPUT_TABLES_DIR)
```

# Cell 3 — Baca full log, compact log, dan trace fitness

```python
FULL_LOG_PATH = PROCESSED_DIR / "04_full_log_process_ready.csv"
COMPACT_LOG_PATH = PROCESSED_DIR / "04_compact_log_process_ready.csv"
TRACE_FITNESS_PATH = PROCESSED_DIR / "07_trace_fitness_main_learning.csv"

full_log = pd.read_csv(FULL_LOG_PATH, parse_dates=["timestamp"])
compact_log = pd.read_csv(COMPACT_LOG_PATH, parse_dates=["timestamp"])
trace_fitness = pd.read_csv(TRACE_FITNESS_PATH)

print("Full log:", full_log.shape)
print("Compact log:", compact_log.shape)
print("Trace fitness:", trace_fitness.shape)

display(full_log.head())
display(compact_log.head())
display(trace_fitness.head())
```

# Cell 4 — Validasi data

```python
print("Jumlah mahasiswa full log:", full_log["case_id"].nunique())
print("Jumlah mahasiswa compact log:", compact_log["case_id"].nunique())
print("Jumlah mahasiswa trace fitness:", trace_fitness["case_id"].nunique())

print("\nJumlah activity full log:", full_log["activity_mapped"].nunique())
print("Jumlah activity compact log:", compact_log["activity_mapped"].nunique())

print("\nDistribusi label performance di full log:")
display(full_log["label_performance"].value_counts())
```

# Cell 5 — Buat data dasar mahasiswa

```python
student_base = (
    full_log[[
        "case_id",
        "nim",
        "nama",
        "kelas",
        "nilai_total",
        "nilai_indeks",
        "label_performance"
    ]]
    .drop_duplicates()
    .sort_values("case_id")
    .reset_index(drop=True)
)

print("Ukuran student_base:", student_base.shape)
display(student_base.head())

print("\nDistribusi label mahasiswa:")
display(student_base["label_performance"].value_counts())
```

# Cell 6 — Fitur jumlah event dan aktivitas unik

```python
basic_features_full = (
    full_log
    .groupby("case_id")
    .agg(
        total_events_full=("activity_mapped", "count"),
        unique_activity_full=("activity_mapped", "nunique")
    )
    .reset_index()
)

basic_features_compact = (
    compact_log
    .groupby("case_id")
    .agg(
        total_events_compact=("activity_mapped", "count"),
        unique_activity_compact=("activity_mapped", "nunique")
    )
    .reset_index()
)

basic_features = basic_features_full.merge(
    basic_features_compact,
    on="case_id",
    how="left"
)

basic_features["compression_ratio"] = (
    basic_features["total_events_compact"] / basic_features["total_events_full"]
).round(4)

display(basic_features.head())

print("Ringkasan basic features:")
display(basic_features.describe().round(2))
```

# Cell 7 — Fitur waktu aktivitas

```python
full_log["date"] = full_log["timestamp"].dt.date

time_features = (
    full_log
    .groupby("case_id")
    .agg(
        first_timestamp=("timestamp", "min"),
        last_timestamp=("timestamp", "max"),
        active_days=("date", "nunique")
    )
    .reset_index()
)

time_features["activity_span_days"] = (
    (time_features["last_timestamp"] - time_features["first_timestamp"]).dt.days + 1
)

time_features["avg_events_per_active_day"] = (
    basic_features_full.set_index("case_id")["total_events_full"]
    / time_features.set_index("case_id")["active_days"]
).values

time_features["avg_events_per_active_day"] = time_features["avg_events_per_active_day"].round(2)

display(time_features.head())

print("Ringkasan time features:")
display(time_features[[
    "active_days",
    "activity_span_days",
    "avg_events_per_active_day"
]].describe().round(2))
```

# Cell 8 — Fitur frekuensi kategori aktivitas

```python
category_counts = (
    full_log
    .groupby(["case_id", "activity_category"])
    .size()
    .unstack(fill_value=0)
    .reset_index()
)

category_counts = category_counts.rename(columns={
    "course": "freq_course",
    "material": "freq_material",
    "quiz": "freq_quiz",
    "assignment": "freq_assignment",
    "forum": "freq_forum"
})

expected_category_cols = [
    "freq_course",
    "freq_material",
    "freq_quiz",
    "freq_assignment",
    "freq_forum"
]

for col in expected_category_cols:
    if col not in category_counts.columns:
        category_counts[col] = 0

category_counts = category_counts[["case_id"] + expected_category_cols]

display(category_counts.head())
```

# Cell 9 — Rasio kategori aktivitas

```python
category_ratio = category_counts.copy()

category_ratio = category_ratio.merge(
    basic_features_full,
    on="case_id",
    how="left"
)

for col in expected_category_cols:
    ratio_col = col.replace("freq_", "ratio_")
    category_ratio[ratio_col] = (
        category_ratio[col] / category_ratio["total_events_full"]
    ).round(4)

category_ratio = category_ratio[[
    "case_id",
    "ratio_course",
    "ratio_material",
    "ratio_quiz",
    "ratio_assignment",
    "ratio_forum"
]]

display(category_ratio.head())
```

# Cell 10 — Fitur aktivitas spesifik

```python
activity_counts = (
    full_log
    .groupby(["case_id", "activity_mapped"])
    .size()
    .unstack(fill_value=0)
    .reset_index()
)

activity_feature_mapping = {
    "Course Viewed": "act_course_viewed",
    "Material Viewed": "act_material_viewed",
    "Quiz Module Viewed": "act_quiz_module_viewed",
    "Quiz Access Prevented": "act_quiz_access_prevented",
    "Quiz Started": "act_quiz_started",
    "Quiz Viewed": "act_quiz_viewed",
    "Quiz Updated": "act_quiz_updated",
    "Quiz Auto-saved": "act_quiz_auto_saved",
    "Quiz Summary Viewed": "act_quiz_summary_viewed",
    "Quiz Submitted": "act_quiz_submitted",
    "Assignment Viewed": "act_assignment_viewed",
    "Assignment Status Viewed": "act_assignment_status_viewed",
    "Forum Viewed": "act_forum_viewed"
}

activity_specific_features = pd.DataFrame()
activity_specific_features["case_id"] = activity_counts["case_id"]

for original_activity, feature_name in activity_feature_mapping.items():
    if original_activity in activity_counts.columns:
        activity_specific_features[feature_name] = activity_counts[original_activity]
    else:
        activity_specific_features[feature_name] = 0

display(activity_specific_features.head())
```

# Cell 11 — Fitur transisi dari compact log

```python
compact_sorted = compact_log.sort_values(
    by=["case_id", "timestamp"]
).copy()

compact_sorted["next_activity"] = (
    compact_sorted
    .groupby("case_id")["activity_mapped"]
    .shift(-1)
)

transitions_student = compact_sorted.dropna(subset=["next_activity"]).copy()

transition_basic_features = (
    transitions_student
    .groupby("case_id")
    .agg(
        total_transition=("next_activity", "count")
    )
    .reset_index()
)

unique_transition_features = (
    transitions_student
    .assign(
        transition_pair=transitions_student["activity_mapped"] + " -> " + transitions_student["next_activity"]
    )
    .groupby("case_id")
    .agg(
        unique_transition_count=("transition_pair", "nunique")
    )
    .reset_index()
)

transition_features = transition_basic_features.merge(
    unique_transition_features,
    on="case_id",
    how="left"
)

display(transition_features.head())

print("Ringkasan transition features:")
display(transition_features.describe().round(2))
```

# Cell 12 — Fitur transisi spesifik utama

```python
transitions_student["transition_pair"] = (
    transitions_student["activity_mapped"] + " -> " + transitions_student["next_activity"]
)

transition_counts = (
    transitions_student
    .groupby(["case_id", "transition_pair"])
    .size()
    .unstack(fill_value=0)
    .reset_index()
)

important_transitions = {
    "Course Viewed -> Material Viewed": "tr_course_to_material",
    "Course Viewed -> Quiz Module Viewed": "tr_course_to_quiz_module",
    "Material Viewed -> Course Viewed": "tr_material_to_course",
    "Quiz Module Viewed -> Course Viewed": "tr_quiz_module_to_course",
    "Quiz Module Viewed -> Quiz Viewed": "tr_quiz_module_to_quiz_viewed",
    "Quiz Module Viewed -> Quiz Started": "tr_quiz_module_to_quiz_started",
    "Quiz Started -> Quiz Viewed": "tr_quiz_started_to_quiz_viewed",
    "Quiz Viewed -> Quiz Updated": "tr_quiz_viewed_to_quiz_updated",
    "Quiz Updated -> Quiz Viewed": "tr_quiz_updated_to_quiz_viewed",
    "Quiz Updated -> Quiz Auto-saved": "tr_quiz_updated_to_auto_saved",
    "Quiz Auto-saved -> Quiz Viewed": "tr_auto_saved_to_quiz_viewed",
    "Quiz Updated -> Quiz Summary Viewed": "tr_quiz_updated_to_summary",
    "Quiz Summary Viewed -> Quiz Submitted": "tr_summary_to_submitted",
    "Quiz Submitted -> Quiz Module Viewed": "tr_submitted_to_quiz_module"
}

transition_specific_features = pd.DataFrame()
transition_specific_features["case_id"] = transition_counts["case_id"]

for transition_name, feature_name in important_transitions.items():
    if transition_name in transition_counts.columns:
        transition_specific_features[feature_name] = transition_counts[transition_name]
    else:
        transition_specific_features[feature_name] = 0

display(transition_specific_features.head())
```

# Cell 13 — Fitur variant dari compact log

```python
variant_df = (
    compact_log
    .sort_values(["case_id", "timestamp"])
    .groupby("case_id")["activity_mapped"]
    .apply(lambda activities: " -> ".join(activities))
    .reset_index(name="variant_sequence")
)

variant_counts = (
    variant_df["variant_sequence"]
    .value_counts()
    .reset_index()
)

variant_counts.columns = ["variant_sequence", "variant_frequency"]
variant_counts["variant_id"] = range(1, len(variant_counts) + 1)

variant_features = variant_df.merge(
    variant_counts,
    on="variant_sequence",
    how="left"
)

variant_features = variant_features[[
    "case_id",
    "variant_id",
    "variant_frequency",
    "variant_sequence"
]]

display(variant_features.head())

print("Jumlah variant unik:", variant_features["variant_id"].nunique())
print("Top 10 variant:")
display(variant_counts.head(10))
```

# Cell 14 — Rapikan trace fitness

```python
trace_fitness_features = trace_fitness[[
    "case_id",
    "trace_is_fit",
    "trace_fitness",
    "missing_tokens",
    "remaining_tokens",
    "consumed_tokens",
    "produced_tokens"
]].copy()

trace_fitness_features["trace_is_fit"] = trace_fitness_features["trace_is_fit"].astype(int)

display(trace_fitness_features.head())
```

# Cell 15 — Gabungkan semua fitur

```python
feature_df = student_base.copy()

feature_tables = [
    basic_features,
    time_features,
    category_counts,
    category_ratio,
    activity_specific_features,
    transition_features,
    transition_specific_features,
    variant_features,
    trace_fitness_features
]

for table in feature_tables:
    feature_df = feature_df.merge(
        table,
        on="case_id",
        how="left"
    )

print("Ukuran feature_df:", feature_df.shape)
display(feature_df.head())
```

# Cell 16 — Cek missing value

```python
missing_features = feature_df.isna().sum()
missing_features = missing_features[missing_features > 0].sort_values(ascending=False)

print("Jumlah kolom dengan missing value:", len(missing_features))
display(missing_features)
```

# Cell 17 — Isi missing numerik dan cek tipe data

```python
numeric_cols = feature_df.select_dtypes(include=["number"]).columns

feature_df[numeric_cols] = feature_df[numeric_cols].fillna(0)

print("Missing value setelah fill numeric:")
display(feature_df.isna().sum()[feature_df.isna().sum() > 0])

print("\nTipe data:")
display(feature_df.dtypes)
```

# Cell 18 — Simpan dataset fitur

```python
FEATURE_OUTPUT_PATH = PROCESSED_DIR / "08_features_process_mining_per_student.csv"
FEATURE_TABLE_OUTPUT_PATH = OUTPUT_TABLES_DIR / "08_features_process_mining_per_student.csv"

feature_df.to_csv(FEATURE_OUTPUT_PATH, index=False)
feature_df.to_csv(FEATURE_TABLE_OUTPUT_PATH, index=False)

print("Dataset fitur berhasil disimpan ke:")
print(FEATURE_OUTPUT_PATH)
print(FEATURE_TABLE_OUTPUT_PATH)
```

# Cell 19 — Ringkasan fitur berdasarkan label

```python
summary_by_label = (
    feature_df
    .groupby("label_performance")
    .agg(
        jumlah_mahasiswa=("case_id", "count"),
        avg_total_events=("total_events_full", "mean"),
        avg_active_days=("active_days", "mean"),
        avg_freq_quiz=("freq_quiz", "mean"),
        avg_freq_material=("freq_material", "mean"),
        avg_freq_assignment=("freq_assignment", "mean"),
        avg_unique_activity=("unique_activity_full", "mean"),
        avg_unique_transition=("unique_transition_count", "mean")
    )
    .round(2)
    .reset_index()
)

display(summary_by_label)

SUMMARY_BY_LABEL_PATH = OUTPUT_TABLES_DIR / "08_summary_features_by_label.csv"
summary_by_label.to_csv(SUMMARY_BY_LABEL_PATH, index=False)

print("Ringkasan fitur per label berhasil disimpan ke:")
print(SUMMARY_BY_LABEL_PATH)
```

## Kesimpulan Feature Extraction

Pada tahap ini, event log hasil process mining diubah menjadi fitur proses pada tingkat mahasiswa. Dataset fitur akhir terdiri dari 89 mahasiswa dengan 65 kolom, yang mencakup identitas mahasiswa, label performa akademik, fitur jumlah event, fitur waktu aktivitas, frekuensi kategori aktivitas, rasio kategori aktivitas, aktivitas spesifik, transisi aktivitas, variant, serta hasil conformance checking.

Hasil ekstraksi fitur menunjukkan bahwa kelompok mahasiswa dengan performa Tinggi cenderung memiliki rata-rata jumlah event yang lebih besar dibandingkan kelompok Sedang dan Rendah. Kelompok Tinggi juga memiliki rata-rata hari aktif, frekuensi aktivitas kuis, frekuensi akses material, jumlah aktivitas unik, dan jumlah transisi unik yang lebih tinggi. Hal ini menunjukkan bahwa mahasiswa dengan performa lebih tinggi memiliki kecenderungan interaksi LMS yang lebih intensif dan lebih bervariasi.

Selain itu, hasil variant analysis menunjukkan bahwa terdapat 89 variant unik dari 89 mahasiswa. Temuan ini menunjukkan bahwa setiap mahasiswa memiliki urutan aktivitas belajar yang berbeda. Dengan demikian, proses belajar mahasiswa pada LMS bersifat sangat bervariasi dan tidak mengikuti satu alur yang sama secara kaku.

Hasil conformance checking berupa trace fitness tetap disimpan sebagai bagian dari fitur proses. Namun, karena seluruh trace memiliki nilai fitness 1,0, fitur tersebut tidak membedakan kelompok performa akademik. Oleh karena itu, fitur yang lebih informatif untuk tahap machine learning adalah fitur frekuensi aktivitas, durasi aktivitas, kategori aktivitas, dan pola transisi aktivitas.

Dataset fitur akhir disimpan sebagai `data/processed/08_features_process_mining_per_student.csv` dan akan digunakan pada tahap analisis pola performa serta pemodelan machine learning.
