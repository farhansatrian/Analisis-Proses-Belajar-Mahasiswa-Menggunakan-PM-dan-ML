# Cell 3 — Fungsi baca tabel

```python
def read_table(path):
    path = Path(path)

    if path.exists():
        print("Loaded:", path.name)
        return pd.read_csv(path)

    print("File tidak ditemukan:", path)
    return None
```

## Kesimpulan Final Result Summary

Notebook ini merangkum seluruh hasil utama dari pipeline process mining dan machine learning. Hasil process mining menunjukkan bahwa proses belajar mahasiswa pada LMS didominasi oleh aktivitas kuis, terutama siklus antara Quiz Viewed dan Quiz Updated. Hasil process discovery dengan Inductive Miner menunjukkan bahwa model proses memiliki kompleksitas karena banyaknya variasi dan pengulangan aktivitas mahasiswa.

Conformance checking menunjukkan bahwa model mampu merepresentasikan seluruh trace pada main learning log, tetapi memiliki precision yang rendah karena model masih fleksibel. Feature extraction dan analisis berdasarkan performa menunjukkan bahwa mahasiswa dengan performa Tinggi cenderung memiliki intensitas aktivitas, hari aktif, akses material, aktivitas unik, dan transisi unik yang lebih tinggi.

Pada tahap machine learning, Random Forest Balanced dengan seluruh fitur process mining memberikan performa data uji terbaik, sedangkan model optimized memberikan performa cross-validation yang lebih stabil. Secara keseluruhan, hasil penelitian menunjukkan bahwa fitur hasil process mining dapat digunakan untuk memahami dan memprediksi performa akademik mahasiswa, meskipun hasil model tetap perlu diinterpretasikan secara hati-hati karena jumlah data terbatas.
